```python
import pandas as pd
import random

# Base feedback templates to ensure variety
pos = ["Great session!", "Very helpful.", "Trainer was excellent.", "Loved the hands-on part.", "Highly recommend."]
neu = ["It was okay.", "Pacing was average.", "Content was as expected.", "Room was fine.", "Standard training."]
neg = ["Too slow.", "Very boring.", "The venue was cold.", "I didn't learn much.", "Waste of time."]

# Generate 100 rows of data
data = []
for i in range(100):
    category = random.choice(['Positive', 'Neutral', 'Negative'])
    if category == 'Positive':
        text = random.choice(pos)
    elif category == 'Neutral':
        text = random.choice(neu)
    else:
        text = random.choice(neg)

    data.append(text)

# Save to CSV (which you can then import into your Google Sheet)
df = pd.DataFrame(data, columns=["Feedback_Text"])
df.to_csv("training_feedback_100.csv", index=False)
print("100 Feedback entries saved to training_feedback_100.csv")
```
