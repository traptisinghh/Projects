```python
## Agent - Arxiv.org
!pip install transformers arxiv
from transformers import pipeline
import arxiv
# Load a Huggingface model for text generation
agent = pipeline(task='text-generation', model = "EleutherAI/gpt-neo-1.3B")
# Functions to search arxiv for relavent papers
def search_arxiv(query, max_results=3):
    search = arxiv.Search(
        query=query,
        max_results=max_results,
        sort_by=arxiv.SortCriterion.Relevance
    )
    return [(result.title, result.pdf_url) for result in search.results()]
```

```python
# Define a search query
query = 'reinforcement learning applications in robotics'
# Retrieve relavnant papers from arxiv
papers = search_arxiv(query)
# Generate a short summary for an AI agent
prompt = f"Summarize recent research on {query} based on these papers: {papers}"
response = agent(prompt, max_length=150)
```

```python
# Display results
print("Relevant arXiv Papers:")
for title, link in papers:
    print(f"- {title}: {link}")
print("\nAI-generated summary:")
print(response[0]['generated_text'])
```
